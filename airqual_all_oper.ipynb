```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("AirQuality.csv")
df
```

```python
df.isnull().sum()
```

```python
## DATA CLEANING ##

df['Ozone']=df['Ozone'].replace(to_replace = np.NaN, value = '12')
```

```python
df
```

```python
df['Solar.R']=df['Solar.R'].replace(to_replace = np.NaN, value = '14')
```

```python
df
```

```python
df.isnull().sum()  //no NULL VALUES
```

```python
## DATA INTEGRATION ##

df1=df[['Ozone','Solar.R']].loc[0:15]
df1
```

```python
df2=df[['Ozone','Solar.R']].loc[16:30]
df2
```

```python
a= pd.concat([df1,df2])
a
```

```python
## ERROR CORRECTING ##

df[df['Month']==4]
```

```python
## MODEL BUILDING ##

from sklearn import preprocessing
lb = preprocessing.LabelEncoder()
df['Month'] = lb.fit_transform(df['Month'])
df
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.iloc[:,:-1], df.iloc[:,-1], test_size = 0.3, random_state = 30)
```

```python
x_train.shape, x_test.shape, y_train.shape
```

```python
from sklearn.naive_bayes import GaussianNB
g = GaussianNB()
g.fit(x_train, y_train)
```

```python
y_pred = g.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
print("Acc : {0:0.4f}", format(accuracy_score(y_test, y_pred)))
```
