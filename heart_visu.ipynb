```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("heart (1).csv")
df
```

```python
df.isnull().sum()
```

```python
a = df.boxplot(column='RestBP', by='RestECG', figsize=(8,4))
a.set_xlabel('RestBP')
a.set_ylabel('RestECG')
a.set_title('BOXPLOT FOR BP AND ECG')
a
```

```python


a = sns.heatmap(df.corr(), cmap='coolwarm', annot=True, square=True, )
plt.title('Correlation')
plt.show(a)
```

```python
df[[i for i in df.columns if i not in ['Age','RestBP']]].hist(bins=30, figsize(15,10))
plt.show()
```

```python
sns.lineplot(x='Age',y='RestBP', data = df)
plt.show()
```

```python
sns.scatterplot(x='Age',y='RestBP', data=df)
plt.show()
```

```python
sns.barplot(x = df['Age'], y = df['Sex'])
```

```python
plt.bar(df['Age'] , df['Sex'], color='red', width=0.5)
plt.xlabel('Age')
plt.ylabel('Sex')
plt.title('AGE VS SEX')
plt.show()
```
